```python
target_map = {'Yes':1, 'No':0}
df["Attrition_numerical"] = df["Attrition"].apply(lambda x: target_map[x])
```

```python
df
```

```python
df= df.drop(['Attrition_numerical'], axis=1)

categorical = []
for col, value in df.iteritems():
    if value.dtype == 'object':
        categorical.append(col)

# Store the numerical columns in a list numerical
numerical = df.columns.difference(categorical)
```

```python
numerical
```

```python
categorical
```

```python
df_new = df[categorical]
```

```python
df_new
```

```python
df_new = df_new.drop(['Attrition'], axis=1) 
```

```python
df_new
```

```python
df_new = pd.get_dummies(df_new)
```

```python
df_new
```

```python
df_num =df[numerical]
```

```python
df_final = pd.concat([df_num, df_new], axis=1)
```

```python
df_final
```

```python
df_final.dtypes
```

```python
target_map = {'Yes':1, 'No':0}
target = df["Attrition"].apply(lambda x: target_map[x])
target.head(3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(df_final,target,test_size = 0.25, random_state= 0)
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=1000,min_samples_split=80,max_depth=7,max_features="sqrt",min_samples_leaf=2)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn import tree, metrics

count_misclassified = (y_test != y_pred).sum()
print('Misclassified samples: {}'.format(count_misclassified))
accuracy = metrics.accuracy_score(y_test, y_pred)
print('Accuracy: {:.2f}'.format(accuracy))
```

```python
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import log_loss
```

```python
confusion_matrix(y_test, y_pred)
```

```python
a= pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
a
```

```python
f1_score(y_test, y_pred, average=None)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
precision = precision_score(y_test, y_pred)
print('Precision: %f' % precision)
recall = recall_score(y_test, y_pred)
print('Recall: %f' % recall)
f1 = f1_score(y_test, y_pred)
print('F1 score: %f' % f1)
```
